# Data preprocessing

```python
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
from nltk.stem.porter import PorterStemmer
from nltk.stem import LancasterStemmer
import re
```

```python
RegTok = RegexpTokenizer("[\w']{3,}")
english_stops = set(stopwords.words('english'))
```

```python
df_posts = df['posts']
```

```python
# PorterStemmer
def tokenizer_1(text):
    tokens = RegTok.tokenize(text.lower()) # 토큰화
    words = [word for word in tokens if (word not in english_stops) and len(word) > 2] # 불용어 제거
    features = (list(map(lambda token: PorterStemmer().stem(token), words))) # 스테밍
    return features
```

```python
df_1 = df_posts.apply(tokenizer_1)
```

```python
df_1[0]
```

```python
# LancasterStemmer
def tokenizer_2(text):
    tokens = RegTok.tokenize(text.lower()) #토큰화
    words = [word for word in tokens if (word not in english_stops) and len(word) > 2] # 불용어 제거
    features = (list(map(lambda token: LancasterStemmer().stem(token), words))) # 스테밍
    return features
```

```python
df_2 = df_posts.apply(tokenizer_2)
```
