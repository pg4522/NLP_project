# Tokenizing

```python
tokenizer = Tokenizer(oov_token="<OOV>")
tokenizer.fit_on_texts(datas['posts'])
```

```python
tokenizer.index_word
```

```python
datas['tokenize'] = datas.apply(lambda row: tokenizer.texts_to_sequences(row['posts']), axis = 1)
datas
```
